```python
import skimage.io as io
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.load('olivetti_faces.npy')
data=data.reshape((data.shape[0],data.shape[1]*data.shape[2]))
```

```python
plt.imshow(data[1].reshape((64,64)),cmap='gray')
plt.axis('off')
```

```python
mean = np.average(data, axis=0)
plt.imshow(mean.reshape((64,64)), cmap='gray')
plt.axis('off')
```

```python
data=data-mean
```

```python
data[0].shape
```

```python
plt.imshow(data[1].reshape((64,64)),cmap='gray')
```

```python
sigma=1/data.shape[0]*np.matmul(np.transpose(data),data)
U,S,V=np.linalg.svd(sigma)
U.shape
```

```python
k=100
U_reduce=U[:,:k]
Z=np.matmul(data,U_reduce)
```

```python
rebuild_data=np.matmul(Z,np.transpose(U_reduce))
```

```python
plt.imshow(rebuild_data[1].reshape((64,64)),cmap='gray')
```

```python
plt.imshow(data[1].reshape((64,64)),cmap='gray')
```

```python
U_reduce.shape
rr=np.transpose(U_reduce)
fig = plt.figure(figsize=(100,100))
for i in range(1,20):
    img = rr[i-1].reshape((64,64))
    fig.add_subplot(10, 2, i)
    plt.imshow(img,cmap='gray')
    plt.axis('off')
plt.show()
```
